```
import matplotlib.pyplot as plt
```

```
x = [2.9,9.0,7,3]
y = [7,15,6,.64]
plt.xlabel("Distance")
plt.ylabel("Time")
plt.plot(x,y,label="track")  #this will draw a line
plt.scatter(x,y,s=177,label="station",marker='*', color='red')   #s for larger the points
plt.grid(color='orange')
plt.legend()    #it will show label in plot
plt.bar(x,y,label="height")
plt.show()
```

```
#Another example of data visualization
player = ['dhoni', 'virat', 'sehwag', 'jadeja']
score = [120, 140, 180, 190]

plt.xlabel("player")
plt.ylabel("score")
#plt.bar(player,score,label="2020 score of Indian Cricket team")
plt.pie(score,labels=player,explode=(0.5,0,0,0),shadow=True,autopct="%2.2ff%%")
plt.grid(color = 'green')
plt.show()
```
