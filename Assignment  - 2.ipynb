```python
# KDE Plot Smoothens out even if there are no values for a value
def kdeplot(feature):
    plt.figure(figsize=(9, 4))
    plt.title("KDE Plot for {}".format(feature))
    ax0 = sns.kdeplot(df[df['trip_duration'] == 0][feature].dropna(), color= 'dodgerblue', label= 'Trip - 0')
    ax1 = sns.kdeplot(df[df['trip_duration'] == 1][feature].dropna(), color= 'orange', label= 'Trip - 1')

kdeplot('passenger_count')
```
